````python
import numpy as np
import pandas as pd
import re

df1 = pd.read_csv("recweekdata.csv")
df2 = pd.read_csv("studentprofiles.csv")
df = pd.merge(df1, df2, on=['firstname', 'lastname']) #merging the two files based on firstname and lastname

#Cleaning birthmonths
#Note: December, November, and October must come first to avoid two or more commands applying to the same code
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"12|Dec$","December",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"11|Nov$","November",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"10|Oct$","October",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"9|Sep$","September",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"8|Aug$","August",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"7|Jul$","July",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"6|Jun$","June",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"5","May",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"4|Apr$","April",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"3|Mar$","March",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"2|Feb$","February",regex=True)
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].str.replace(r"1|Jan$","January",regex=True)

#Cleaning courses
#Note: order must follow BS ME>BS MGT>BS LM otherwise code doesn't work
df.loc[:,"course"] = df.loc[:,"course"].str.replace(r"Management Engineering|BS ME|ME","BS Management Engineering",regex=True)
df.loc[:,"course"] = df.loc[:,"course"].str.replace(r"Management|BS MGT|MGT","BS Management",regex=True)
df.loc[:,"course"] = df.loc[:,"course"].str.replace(r"Legal Management|BS LM|LM","BS Legal Management",regex=True)
df.loc[:,"course"] = df.loc[:,"course"].str.replace(r"Communications Technology|BS CTM|CTM","BS Communications Technology Management",regex=True)
df.loc[:,"course"] = df.loc[:,"course"].str.replace(r"Information Technology Entrepreneurship|BS ITE|ITE","BS Information Technology Entrepreneurship",regex=True)

#Cleaning phonenumbers
df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].str.replace(r"\+63 9",r"09", regex=True)
df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].str.replace(r"^9",r"09", regex=True)

#Cleaning birthyears
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^(9[7-9])",r"19\1", str(x)))

#Identifying the carrier
carrier = {"0917":"Globe","0918":"Smart", "0927":"Globe"}
df.loc[:,"phonenumber"].str[0:4].apply(lambda x: carrier[x])
df["carrier"] = df.loc[:,"phonenumber"].str[0:4].apply(lambda x: carrier[x])
````
